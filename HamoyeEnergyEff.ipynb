```python
import pandas as pd
import math
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/Hamoye/energydata_complete.csv')
```

```python
model = LinearRegression()

xtt, xts, ytr, yts = train_test_split(df[['T2']], df[['T6']], test_size=0.25)

model.fit(xtt, ytr)
```

```python
pred = model.predict(xts)
```

```python
rmse = math.sqrt(mean_squared_error(yts, pred))
print(rmse)
```

Multiple regression :

```python
tgt = df[['Appliances']]
newdf = df.drop(['date', 'lights','Appliances' ], axis=1)
```

```python
xtt, xts, ytr, yts = train_test_split(newdf, tgt, test_size=0.3, random_state=42)
```

```python
scaler = MinMaxScaler()
scaler.fit_transform(xtt)
scaler.fit_transform(xts)
```

```python
model.fit(xtt, ytr)

trpred = model.predict(xtt)
pred = model.predict(xts)
```

```python
trme = mean_absolute_error(ytr, trpred)
mae = mean_absolute_error(yts, pred)
print(trme)
print(mae)
```

```python
trmse = math.sqrt(mean_squared_error(ytr, trpred))
print(trmse)
rmse = math.sqrt(mean_squared_error(yts, pred))
print(rmse)
```

Lasso

```python
lasso_model = Lasso()
lasso_model.fit(xtt, ytr)

pred = lasso_model.predict(xts)
feature_weights = lasso_model.coef_

non_zero_features = sum(feature_weights != 0)

print("Number of features with non-zero weights:", non_zero_features)

rmse = math.sqrt(mean_squared_error(yts, pred))
print(rmse)
```

Ridge

```python
ridge_model = Ridge()
ridge_model.fit(xtt, ytr)

y_pred_test = ridge_model.predict(xts)

rmse_test = np.sqrt(mean_squared_error(yts, y_pred_test))

print("Root Mean Squared Error (RMSE) on test set with Ridge regression:", rmse_test)
```
